### Merge the two dataframes

```python
df_merge = pd.merge(train, test, how = "outer")
```

```python
df_merge
```

### Performing Join on the two dataframes

```python
df_join = train.join(test, how = "outer")
```

#### We cannot perform a join operation on the two dataframes because both dataframes have the same set of columns which would cause an overlap while joining them.

#### Joining is better used when combining the columns of two differently-indexed dataframes into a single dataframe.
